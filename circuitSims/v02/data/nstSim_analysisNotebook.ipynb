```python
import numpy as np
import math as mp

import os
import csv
from os import listdir
from os.path import isfile, join

import matplotlib as mpl
import matplotlib.pyplot as plt

def bestSaveFig(name):
    plt.savefig(name, bbox_inches = 'tight', pad_inches = 0.1, dpi = 500)

def getFileNames(path):
    return([f for f in listdir(path) if isfile(join(path, f))])

def spiceAmpExtracter(splitOneOne):
    splitTwo = (splitOneOne).split("(")
    splitThree = (splitTwo[1]).split(",")
    splitFour = (splitThree[0]).split("d")
    return(float(splitFour[0]))

def readFile(file):
    stream = open(path + '/' + file, mode='r')
    firstLine = stream.readline()
    
    freqVec = []
    realVec = []
    imagVec = []
    
    for line in stream:
        splitOne = line.split('\t')
        freq = float(splitOne[0])
        splitThree = (splitOne[1]).split(",")
        real = float(splitThree[0])
        imag = float(((splitThree[1]).split("\\"))[0])
        
        freqVec.append(freq)
        realVec.append(real)
        imagVec.append(imag)
    
    magnVec = np.sqrt(np.add(np.square(realVec), np.square(imagVec)))
    
    return(freqVec, realVec, imagVec, magnVec)
```

```python
path = "lopper"

paraFiles = getFileNames(path)

for file in paraFiles:
    freqVec, realVec, imagVec, magnVec = readFile(file)
    plt.semilogx(freqVec, np.multiply(imagVec, 1e12), label = "left="+file[17]+"mm")
    
plt.legend()
plt.grid()
plt.xlabel(r"Frequency [Hz]", size = 15)
plt.ylabel(r"Current [pA]", size = 15)
#bestSaveFig("lopsidedSims_curr-vs-freq_17apr2025.png")
plt.show()
```
